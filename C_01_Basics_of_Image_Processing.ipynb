### Computer Vision

- **Alternate Name**: Also known as machine vision.

- **Definition**: Refers to the process of extracting information from images, text, videos, and other visual data.

- **Objective**: Aims to replicate human vision.

- **Relationship to AI**: As a subset of artificial intelligence, computer vision collects data from digital images and videos to analyze and identify various attributes.

### Applications

### Understanding basics Computer Vision

1. **Pixels**:  
   - The smallest unit of a digital image is the pixel
   - short form for pixel element
   - contains color information (usually represented in RGB or grayscale).
   - square or rectangular dot that represents a single point of colour and intensity wihin an image 
   - Pixels are arranged in a grid pattern 
   - size and no:of pixels in an image determines its resolution 
   - Image resolution is defined by the width and height in pixels (e.g., 1920x1080).
   - Higher resolution (more pixels) generally leads to better detail and clarity, while lower resolution can result in a loss of visual fidelity.

2. **Image Formats**:  
   - Common image formats include:
     - **JPEG**: Lossy compression; great for photographs.
     - **PNG**: Lossless compression; supports transparency.
     - **GIF**: Limited color palette; supports animations.
     - **TIFF**: High-quality, lossless format; often used in professional settings.

3. **Color Models**:  
   - **RGB**: Red, Green, Blue; used for most digital images.
   - **CMYK**: Cyan, Magenta, Yellow, Black; used for print media.
   - **Grayscale**: Represents images in shades of gray, often used in image processing.

4. **Types of Images**:  

      1. **Binary Images**:
         - **Definition**: Images that consist of only two colors: black and white.
         - **Pixel Representation**: Each pixel is represented by a single bit (0 for black and 1 for white).
         - **Usage**: Commonly used in simple graphics, text documents, and for tasks like image segmentation and thresholding in computer vision.

      2. **Grayscale Images**:
         - **Definition**: Images that represent shades of gray, varying from black to white.
         - **Pixel Representation**: Each pixel is typically represented by 8 bits, allowing for 256 different shades (0 for black, 255 for white, with values in between representing varying gray levels).
         - **Usage**: Used in photography, medical imaging, and computer vision, where color information is not necessary but intensity variations are important.

      3. **Color Images**:
         - **Definition**: Images that include a range of colors, typically represented using color models like RGB (Red, Green, Blue).
         - **Pixel Representation**: Each pixel is often represented by 24 bits (8 bits for each color channel), allowing for over 16 million color combinations.
         - **Usage**: Commonly used in digital photography, graphics, and any application where color differentiation is important.

5. **Common Screen Resolutions**:  

      1. **Standard Resolutions**:
            - **480p (SD)**: 640 x 480 pixels
            - **720p (HD)**: 1280 x 720 pixels
            - **1080p (Full HD)**: 1920 x 1080 pixels
            - **1440p (Quad HD)**: 2560 x 1440 pixels
            - **2160p (4K Ultra HD)**: 3840 x 2160 pixels
            - **4320p (8K Ultra HD)**: 7680 x 4320 pixels

      2. **Widescreen Resolutions**:
         - **16:9 Aspect Ratio**:
         - Common in TVs and monitors, includes 720p, 1080p, and 4K.
         - **21:9 Aspect Ratio** (Ultrawide):
         - **UltraWide QHD**: 3440 x 1440 pixels
         - **UltraWide 4K**: 5120 x 2160 pixels

      3. **Mobile Resolutions**:
         - **HD**: 1280 x 720 pixels
         - **Full HD**: 1920 x 1080 pixels
         - **Quad HD**: 2560 x 1440 pixels
         - **4K**: 3840 x 2160 pixels (in premium smartphones)

      4. **Legacy Resolutions**:
         - **XGA**: 1024 x 768 pixels
         - **SVGA**: 800 x 600 pixels
         - **VGA**: 640 x 480 pixels

      5. **Other Resolutions**:
         - **4:3 Aspect Ratio**:
         - Common in older monitors and projectors: 1024 x 768, 800 x 600.
         - **Custom Resolutions**:
         - Some applications and devices may support custom resolutions based on specific needs.

6. **Key Libraries**:

### OpenCV

- **Full Name**: Open Source Computer Vision Library.
- **Overview**: A powerful and widely-used open-source library designed for computer vision and image processing tasks.

- **Purpose**:  
  - Provides tools and functions for developers to create applications that process and analyze images and videos.
  - Supports tasks such as object detection, face recognition, and image transformations.

- **Languages**:  
  - Primarily written in C++.
  - Offers bindings for Python, Java, and other languages, enhancing accessibility for a diverse range of developers.

- **Features**:
  - **Image Processing**: Includes functions for filtering, edge detection, color manipulation, and geometric transformations.
  - **Object Detection**: Provides pre-trained models and algorithms for recognizing and tracking objects in images and videos.
  - **Facial Recognition**: Offers tools for detecting and recognizing faces, important for security and user authentication.
  - **Machine Learning**: Integrates with machine learning algorithms to improve image analysis capabilities.

- **Cross-Platform Compatibility**:  
  - Usable on various operating systems, including Windows, Linux, and macOS.
  - Also compatible with mobile platforms like Android and iOS.

### Pillow

- **Full Name**: Pillow (Python Imaging Library Fork).
- **Overview**: A powerful and user-friendly open-source library for image processing in Python.

- **Purpose**:  
  - Provides tools for opening, manipulating, and saving various image file formats.
  - Supports tasks such as image enhancement, transformation, and filtering.

- **Supported Formats**:  
  - Can handle a wide range of image formats, including JPEG, PNG, GIF, BMP, and TIFF.

- **Features**:
  - **Image Processing**: Includes functions for resizing, cropping, rotating, and transforming images.
  - **Image Enhancement**: Provides tools for adjusting brightness, contrast, color balance, and sharpness.
  - **Drawing**: Offers capabilities for drawing shapes, text, and images onto existing images.
  - **Filters**: Includes a variety of built-in filters for image effects, such as blur and contouring.

- **Ease of Use**:  
  - Designed to be easy to use, making it accessible for beginners and experienced developers alike.

- **Cross-Platform Compatibility**:  
  - Works on multiple operating systems, including Windows, macOS, and Linux.

Loading and Displaying Images

```python
"""Using OpenCV"""
# Load the image using OpenCV
image = cv2.imread('image.jpg')

# Convert color from BGR (OpenCV default) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
"""Using Pillow"""
# Load and display using Pillow
image_pil = Image.open('image.jpg')
plt.imshow(image_pil)
plt.axis('off')
plt.show()
```

Using OpenCV and Matplotlib to Display Different Image Types

```python
import cv2
import matplotlib.pyplot as plt

# Load the original color image
image = cv2.imread('image.jpg')

# Check if the image is loaded successfully
if image is None:
    raise ValueError("Error loading image: image.jpg")

# Convert to Grayscale
grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Convert to Binary Image using a threshold
# You can adjust the threshold value (127) as needed
_, binary_image = cv2.threshold(grayscale_image, 127, 255, cv2.THRESH_BINARY)

# Set up subplots for displaying different image types
plt.figure(figsize=(12, 4))

# Display Original Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))  # Convert to RGB for display
plt.title('Original Image')
plt.axis('off')

# Display Grayscale Image
plt.subplot(1, 3, 2)
plt.imshow(grayscale_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

# Display Binary Image
plt.subplot(1, 3, 3)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')
plt.axis('off')

# Show the plot
plt.tight_layout()
plt.show()
```

Basic Transformations

1. Resizing

```python
resized_image = cv2.resize(image, (200, 200))  # Resize to 200x200 pixels
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
resized_image_pil = image_pil.resize((200, 200))
plt.imshow(resized_image_pil)
plt.axis('off')
plt.show()
```

2. Cropping

```python
# Cropping a 100x100 region from the top left corner
cropped_image = image[0:100, 0:100]
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Crop with Pillow (left, upper, right, lower)
cropped_image_pil = image_pil.crop((0, 0, 300, 300))
plt.imshow(cropped_image_pil)
plt.axis('on')
plt.show()
```

3. Rotation

```python
# Rotate 90 degrees clockwise
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Rotate 45 degrees with Pillow
rotated_image_pil = image_pil.rotate(45)  # Rotates counter-clockwise
plt.imshow(rotated_image_pil)
plt.axis('off')
plt.show()
```

4. Flipping

```python
# Flip vertically
flipped_image = cv2.flip(image, 0)  # 0 for vertical flip, 1 for horizontal flip
plt.imshow(cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Flip horizontally with Pillow
flipped_image_pil = image_pil.transpose(Image.FLIP_LEFT_RIGHT)
plt.imshow(flipped_image_pil)
plt.axis('off')
plt.show()
```
