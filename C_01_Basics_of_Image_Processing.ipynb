Loading and Displaying Images

```python
"""Using OpenCV"""
# Load the image using OpenCV
image = cv2.imread('image.png')

# Convert color from BGR (OpenCV default) to RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
"""Using Pillow"""
# Load and display using Pillow
image_pil = Image.open('image.png')
plt.imshow(image_pil)
plt.axis('off')
plt.show()
```

Basic Transformations

1. Resizing

```python
resized_image = cv2.resize(image, (200, 200))  # Resize to 200x200 pixels
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
resized_image_pil = image_pil.resize((200, 200))
plt.imshow(resized_image_pil)
plt.axis('off')
plt.show()
```

2. Cropping

```python
# Cropping a 100x100 region from the top left corner
cropped_image = image[0:100, 0:100]
plt.imshow(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Crop with Pillow (left, upper, right, lower)
cropped_image_pil = image_pil.crop((0, 0, 300, 300))
plt.imshow(cropped_image_pil)
plt.axis('on')
plt.show()
```

3. Rotation

```python
# Rotate 90 degrees clockwise
rotated_image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Rotate 45 degrees with Pillow
rotated_image_pil = image_pil.rotate(45)  # Rotates counter-clockwise
plt.imshow(rotated_image_pil)
plt.axis('off')
plt.show()
```

4. Flipping

```python
# Flip vertically
flipped_image = cv2.flip(image, 0)  # 0 for vertical flip, 1 for horizontal flip
plt.imshow(cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python
# Flip horizontally with Pillow
flipped_image_pil = image_pil.transpose(Image.FLIP_LEFT_RIGHT)
plt.imshow(flipped_image_pil)
plt.axis('off')
plt.show()
```
